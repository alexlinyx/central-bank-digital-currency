```python
import matplotlib.pyplot as plt
```

```python
def read_file(path):
    data = []
    with open(path, 'r') as f:
        df = f.readlines()
        for line in df:
            data.append(int(line.strip()))
    return data
```

```python
bc1 = read_file('blockchains/output1.txt')
bc2 = read_file('blockchains/output2.txt')
db1 = read_file('database/output1.txt')
db2 = read_file('database/output2.txt')
new1 = read_file('combined/output1.txt')
new2 = read_file('combined/output2.txt')
```

```python
x1 = range(100)
x2 = range(10,100)
```

```python
def plot(y,x, xlabel, title):
    plt.plot(x,y)
    plt.ylabel('Microseconds')
    plt.xlabel(xlabel)
    plt.title(title)
    plt.show()
```

```python
plot(bc1, x1, 'Number of transactions', 'Blockchain')
```

```python
plot(bc2, x2, 'Number of wallets', 'Blockchain')
```

```python
plot(db1, x1, 'Number of transactions', 'Database')
```

```python
plot(db2, x2, 'Number of wallets', 'Database')
```

```python
plot(new1, x1, 'Number of transactions', 'Combined')
```

```python
plot(new2, x2, 'Number of wallets', 'Combined')
```

```python
plt.plot(x1,bc1, label='Vanilla blockchain')
plt.plot(x1,new1, label='With central database')
plt.legend()
plt.xlabel('Number of transactions')
plt.ylabel('Microseconds')
plt.show()
```

```python
plt.plot(x2,bc2, label='Vanilla blockchain')
plt.plot(x2, new2, label='With central database')
plt.legend()
plt.xlabel('Number of wallets')
plt.ylabel('Microseconds')
plt.show()
```
