```python
import matplotlib.pyplot as plt
```

```python
def read_file(path):
    path = 'results/' + path
    data = []
    with open(path, 'r') as f:
        df = f.readlines()
        for line in df:
            data.append(int(line.strip()))
    return data
```

```python
bc1 = read_file('blockchains1.txt')
bc2 = read_file('blockchains2.txt')
db1 = read_file('custom1.txt')
db2 = read_file('custom2.txt')
combined_many1 = read_file('combined_many1.txt')
combined_many2 = read_file('combined_many2.txt')
combined_single1 = read_file('combined_single1.txt')
combined_single2 = read_file('combined_single2.txt')
cloud_many1 = read_file('dynamo_many1.txt')
cloud_many2 = read_file('dynamo_many2.txt')
cloud_single1 = read_file('dynamo_single1.txt')
cloud_single2 = read_file('dynamo_single2.txt')
```

```python
x1 = range(100)
x2 = range(10,100)
```

```python
def plot(y,x, xlabel, title):
    plt.plot(x,y)
    plt.ylabel('Microseconds')
    plt.xlabel(xlabel)
    plt.title(title)
    plt.show()
```

```python
plt.plot(x1,bc1, label='Simple Blockchain')
plt.plot(x1,db1, label='In-memory Hashtable')
plt.plot(x1,combined_many1, label='Multiple Chains w/ In-memory Hashtable')
plt.plot(x1,combined_single1, label='Single Chain w/ In-memory Hashtable')
plt.legend()
plt.xlabel('Number of transactions')
plt.ylabel('Microseconds')
plt.show()
```

```python
plt.plot(x2,bc2, label='Simple Blockchain')
plt.plot(x2,db2, label='In-memory Hashtable')
plt.plot(x2,combined_many2, label='Multiple Chains w/ In-memory Hashtable')
plt.plot(x2,combined_single2, label='Single Chain w/ In-memory Hashtable')
plt.legend()
plt.xlabel('Number of wallets')
plt.ylabel('Microseconds')
plt.show()
```

```python
plt.plot(x1,cloud_many1, label='Multiple Chains w/ DynamoDB')
plt.plot(x1,cloud_single1, label='Single Chain w/ DynamoDB')
plt.legend()
plt.xlabel('Number of transactions')
plt.ylabel('Microseconds')
plt.show()
```

```python
plt.plot(x2,cloud_many2, label='Multiple Chains w/ DynamoDB')
plt.plot(x2,cloud_single2, label='Single Chain w/ DynamoDB')
plt.legend()
plt.xlabel('Number of wallets')
plt.ylabel('Microseconds')
plt.show()
```
